```python
import pandas as pd
```

```python
football_df = pd.read_csv("data.csv")
```

```python
football_df.head()
```

```python
cols_to_drop = [
    "shot_id_number",
    "team_name",
    "home/away",
    "date_of_game",
    "team_id",
    "remaining_min.1",
    "power_of_shot.1",
    "knockout_match.1",
    "remaining_sec.1",
    "distance_of_shot.1",
]
```

```python
football_df.drop(cols_to_drop,inplace=True,axis=1)
```

```python
football_df.shape
```

```python
football_df.head()
```

```python
football_df.drop("Unnamed: 0", inplace=True, axis = 1)
```

```python
football_df.head()
```

```python
football_df.iloc[:,2]
```

```python
from sklearn.preprocessing import Imputer
imputer = Imputer(missing_values = 'NaN', strategy = 'mean', axis = 0)
imputer = imputer.fit(football_df.iloc[:, 1:3])
football_df.iloc[:, 1:3] = imputer.transform(football_df.iloc[:, 1:3])
```

```python
football_df[["lat","lng"]] = football_df["lat/lng"].str.split(",",expand = True)
```

```python
football_df.head()
```

```python
football_df.drop("lat/lng", inplace=True, axis = 1)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()
football_df.iloc[:, 10] = labelencoder.fit_transform(str(football_df.iloc[:, 10]))
onehotencoder = OneHotEncoder(categorical_features = [10])
football_df = onehotencoder.fit_transform(football_df)
```
