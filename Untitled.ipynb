```python
range_map = {}
```

```python
filtered_df = df[df['distance_of_shot'] == df['distance_of_shot.1']]
```

```python
for val in shot_range_vals:
    r1 = filtered_df[filtered_df['range_of_shot'] == val]['distance_of_shot.1']
    r2 = filtered_df[filtered_df['range_of_shot'] == val]['distance_of_shot']
    range_map[val] = range(int(min(r1.min(), r2.min())), int(max(r1.max(), r2.max())))
```

```python
range_map
```

```python
def get_range(distance): 
    for label, r in range_map.items():
        if distance in r:
            return label
```

```python
def update_range(row):
    if not pd.isna(row['range_of_shot']): return row['range_of_shot']
    if row['distance_of_shot'] != row['distance_of_shot.1']:
        if get_range(row['distance_of_shot']):
            return get_range(row['distance_of_shot'])
        else:
            return get_range(row['distance_of_shot.1'])
    else:
        return get_range(row['distance_of_shot'])
```

```python
df['range_of_shot'] = df.apply(update_range, axis=1)
```

```python
area_map = {}
filtered_df = df[~df['area_of_shot'].isnull()]
for val in set(filtered_df['area_of_shot']):
    row = filtered_df[filtered_df['area_of_shot'] == val]
    area_map[val] = \
        range(int(row['location_x'].min()), int(row['location_x'].max())), \
        range(int(row['location_y'].min()), int(row['location_y'].max()))
```

```python
area_map
```

```python
def update_area(row):
    if not pd.isna(row['area_of_shot']): return row['area_of_shot']
    for key, value in area_map.items():
        if not pd.isna(row['location_x']) and not pd.isna(row['location_y']):
            if row['location_x'] in value[0] and row['location_y'] in value[1]:
                return key
        elif not pd.isna(row['location_x']):
            if row['location_x'] in value[0]:
                return key
        elif not pd.isna(row['location_y']):
            if row['location_y'] in value[1]:
                return key
    return row['area_of_shot']
```

```python
df['area_of_shot'] = df.apply(update_area, axis=1)
```

```python
res = df[~df['game_season'].isnull()][['game_season', 'match_id']]
```

```python
match_id_season_map = dict(res.groupby('game_season')['match_id'].apply(list))
```

```python
def updateGameSeason(row):
    if not pd.isna(row['game_season']): return row['game_season']
    for key, value in match_id_season_map.items():
        if row['match_id'] in value:
            return key
    return np.random.choice(list(match_id_season_map.keys()))
```

```python
df['game_season'] = df.apply(updateGameSeason, axis = 1)
```

```python
def update_knockout_match(row):
    if not pd.isna(row['knockout_match']): return row['knockout_match']
    if not pd.isna(row['knockout_match.1']) and int(row['knockout_match.1']) in [0, 1]:
        return row['knockout_match.1']
    return row['knockout_match']
```

```python
df['knockout_match'] = df.apply(update_knockout_match, axis = 1)
```

```python
cnt_0 = 0
cnt_1 = 0
for i, row in df[~df['knockout_match'].isnull()].iterrows():
    if row['knockout_match'] == 0:
        cnt_0 += 1
    elif row['knockout_match'] == 0:
        cnt_1 += 1
```

```python
np.random.choice([0, 1], p = [cnt_0 / (cnt_0+cnt_1), cnt_1 / (cnt_0+cnt_1)])
```

```python
def fill_random(row):
    if not pd.isna(row['knockout_match']): return row['knockout_match']
    
    return np.random.choice([0, 1], p = [cnt_0 / (cnt_0+cnt_1), cnt_1 / (cnt_0+cnt_1)])
```

```python
df['knockout_match'] = df.apply(fill_random, axis = 1)
```
